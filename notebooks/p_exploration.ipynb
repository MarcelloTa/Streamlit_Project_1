```python
# Es wird doch ein Kaggle-Datensatz... CSV laden:
imdb = pd.read_csv(r'../data/imdb.csv')
imdb.head()
```

### Datensatz-Bereinigung und -Anpassungen einiger Datentypen

```python
# Momentane Datentypen überprüfen
imdb.dtypes
```

```python
# Null-Werte und die dazugehörigen Zeilen entfernen
imdb.dropna(inplace=True)
```

```python
# Betroffene Zeilen in die gewünschten Datentypen casten
imdb['Year'] = imdb['Year'].astype(int)
imdb['Duration (min)'] = imdb['Duration (min)'].astype(int)
imdb['Metascore'] = imdb['Metascore'].astype(int)
imdb['Votes'] = imdb['Votes'].str.replace(',', '').astype(int)
imdb['Review Count'] = imdb['Review Count'].str.replace(',', '').astype(int)
```

```python
# Überprüfung der Datentypen
imdb.dtypes
```

```python
# Ergebnisse im DataFrame überprüfen
imdb
```

```python
# Poster-Spalte entfernen
imdb.drop(columns=['Poster'], inplace=True)
```

```python
# Index neu setzen
imdb.reset_index(drop=True, inplace=True)
```

```python
# Überprüfung
imdb
```

```python
# Leerzeichen in den Spaltennamen durch _ ersetzen
imdb.columns = imdb.columns.str.replace(' ', '_')
```

```python
# Für die Genres nur das Hauptgenre behalten
imdb['Genre'] = imdb['Genre'].str.split(',').str[0]
```

```python
# DataFrame auf neue Variable legen
imdb_clean = imdb
imdb_clean.head()
```
